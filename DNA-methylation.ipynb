
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests

# Step 4: Compute mean beta values
mean_group1 = group1_m_values.mean(axis=1)
mean_group2 = group2_m_values.mean(axis=1)

# Step 5: Compute log2 fold change (Case / Control)
epsilon = 1e-6
fold_change = np.log2((mean_group2 + epsilon) / (mean_group1 + epsilon))

# Step 6: T-test per CpG
p_values = []
for i in range(len(m_value)):
    stat, p = ttest_ind(group1_m_values.iloc[i, :], group2_m_values.iloc[i, :], equal_var=False)
    p_values.append(p)

# Step 7: Adjust p-values (FDR)
adj_p_values = multipletests(p_values, method='fdr_bh')[1]

# Step 8: Compile results
result_df = pd.DataFrame({
    'CpG': cpg_ids,
    'mean_control': mean_group1,
    'mean_case': mean_group2,
    'log2FC': fold_change,
    'p_value': p_values,
    'adj_p_value': adj_p_values
})

# Step 9: Apply filtering criteria
filtered = result_df[
    (abs(result_df['log2FC']) >= np.log2(1.5)) & 
    (result_df['p_value'] < 0.01)
]

# Step 10: Save results
filtered.to_csv('D:/Desktop/Dataset/DNA_methylation/DMPs_python_filtered.csv', index=False)
print(f"{len(filtered)} DMPs identified and saved.")
